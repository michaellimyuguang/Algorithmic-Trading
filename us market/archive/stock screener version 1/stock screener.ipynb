```python
import datetime as dt
import pandas as pd
import pandas_datareader.data as web
import numpy as np
```

```python
def indicators_calculation(ticker):
    try:
#         end = dt.datetime.today()
#         start = dt.datetime.today() - dt.timedelta(days=100)

        end = dt.datetime(2021, 3, 17)
        start = end - dt.timedelta(days=100)
        df = web.DataReader(ticker, 'yahoo', start, end)
        
        #20SMA and 40SMA    
        df['20SMA'] = df['Close'].rolling(window=20).mean()
        df['40SMA'] = df['Close'].rolling(window=40).mean()

        #5CCI
        #typical price used for cci calculation
        df['TP'] = (df['High'] + df['Low'] + df['Close']) / 3
        #mean deviation formula used in cci calculation
        MD = lambda x : np.mean(np.abs(x-np.mean(x)))
        #cci calculation
        df['CCI'] = (df['TP'] - df['TP'].rolling(5).mean()) / (0.015 * df['TP'].rolling(5).apply(MD, True)) 
        return df
    
    except Exception:    
        df = pd.DataFrame()
        return df

def indicator_condition(df):
    if (df['20SMA'].iloc[-1] > df['40SMA'].iloc[-1]) and (df['CCI'].iloc[-1] < -100) and (df['Close'].iloc[-1] <= df['20SMA'].iloc[-1] and (df['Close'].iloc[-1] >= df['40SMA'].iloc[-1]):
        return True
```

```python
df_tickers = pd.read_csv("us tickers.csv")
df_tickers = df_tickers.dropna(subset=["Sector"])
symbol_list = []
name_list = []
satisfied_list = []
i = 0

symbol_list = df_tickers['Symbol'].tolist()
name_list = df_tickers['Name'].tolist()
ticker_dict = dict(zip(symbol_list, name_list)) 

total = len(df_tickers)
print('total number of tickers ' + str(total))
for key in ticker_dict:
    i = i + 1
    df = indicators_calculation(key)
    print(len(df))
    if len(df) > 40:
        satisfied = indicator_condition(df)
        
        if satisfied:
            satisfied_list.append(key)
            print(key)
            print(df.tail(2))
    satisfied = False
    print('finished ticker no. ' + str(i))
```
