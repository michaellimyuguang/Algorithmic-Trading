```python
import datetime as dt
import pandas as pd
import pandas_datareader.data as web
```

```python
df_tickers = pd.read_csv(r"C:\Users\michael\Desktop\trading materials\us market\us tickers.csv")
df_tickers = df_tickers.dropna(subset=["Sector"])
df_tickers.reset_index(drop=True, inplace=True)
display(df_tickers)
```

```python
df_tickers = df_tickers[~df_tickers['Symbol'].str.contains("\^|\/")]
df_tickers.reset_index(drop=True, inplace=True)
display(df_tickers)
```

```python
def download_df(ticker):
    try:
        end = dt.datetime(2021, 4, 14)
        start = dt.datetime(2021, 1, 1) - dt.timedelta(days=310)
        df = web.DataReader(ticker, 'yahoo', start, end)
        
        return df
    
    except Exception:
        df = pd.DataFrame()
        
        return df
```

```python
path = r"C:\Users\michael\Desktop\trading materials\us market\us_tickers_history"
for i in range(3546, len(df_tickers)):
    ticker = df_tickers['Symbol'].iloc[i]
    df = download_df(ticker)
    print('no: ' + str(i))
    print(len(df))
    if len(df) >= 280:
        df.to_csv(path + r"\\" + ticker + ".csv")
```
