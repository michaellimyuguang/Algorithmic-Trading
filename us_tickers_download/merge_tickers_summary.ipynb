this notebook is used to merge the information sheet from the 3 exhanges: NASDAQ, NYSE and AMEX

```python
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
mypath = r"C:\Users\yu guang\Desktop\us_tickers_download"
onlyfiles = [f for f in listdir(mypath) if 'nasdaq_screener_' in f]

df = pd.DataFrame()

for i in range(len(onlyfiles)):
    df2 = pd.read_csv(mypath + r"\\" + onlyfiles[i])
    df = df.append(df2, ignore_index=True)

df = df.dropna(subset=["Sector"])
df.reset_index(drop=True, inplace=True)
df = df[~df['Symbol'].str.contains("\^|\/")]
df.reset_index(drop=True, inplace=True)

display(df)
df.to_csv(mypath + r"\\" + "us_tickers_summary.csv", index=False)
```
